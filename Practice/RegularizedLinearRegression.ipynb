```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
from sklearn.linear_model import Ridge
```

```python
# X is a 5*5 Hilbert matrix
X = 1./(np.arange(1,6)+np.arange(0,5).reshape(-1,1))
y = np.ones(5)
```

```python
X
```

```python
alphas = np.logspace(-10,2,200)
coefs = []
for a in alphas:
    ridge = Ridge(alpha=a,fit_intercept=False)
    ridge.fit(X,y)
    coefs.append(ridge.coef_)
    
```

```python
plt.figure(figsize=(10,8))
ax = plt.gca()
ax.plot(alphas,coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficient')
plt.axis('tight')
plt.show()
```

```python
from sklearn import linear_model
```

```python
x = np.random.random(100)*10
y = 0.5*x + np.random.random(100)
lr = linear_model.LinearRegression(normalize=True)
lr.fit(x.reshape(-1,1),y)
y_pred = lr.predict(x.reshape(-1,1))
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
print('Coefficient: '+str(lr.coef_))
```

```python
ix = y.argmax()
y[30] = 100
lr = linear_model.Ridge(alpha=0.1,normalize=True)
lr.fit(x.reshape(-1,1),y)
y_pred = lr.predict(x.reshape(-1,1))
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
print('Coefficient: '+str(lr.coef_))
```
