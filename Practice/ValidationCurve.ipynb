```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model, preprocessing
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
x = np.linspace(-4,4,100).reshape(-1,1)
y = np.cos(x).reshape(-1,1) + np.random.random(x.size).reshape(-1,1)
```

```python
x.shape,y.shape
```

```python
degrees = (1,4,7)
for d in degrees:
    poly_f = preprocessing.PolynomialFeatures(degree=d)
    x_adj = poly_f.fit_transform(x)
    lr_model = linear_model.LinearRegression()
    lr_model.fit(x_adj,y)
    y_pred = lr_model.predict(x_adj)
    plt.figure(figsize=(5,3))
    plt.scatter(x,y)
    plt.plot(x,y_pred, color='orange')
    score = r2_score(y,y_pred)
    plt.title('Score: '+str(score))
    plt.show()
```

```python
depths = (2,4,8)
for d in depths:
    d_tree = DecisionTreeRegressor(max_depth=d)
    d_tree.fit(x,y)
    y_pred = d_tree.predict(x)
    plt.figure(figsize=(5,3))
    plt.scatter(x,y)
    score = r2_score(y,y_pred)
    plt.title('Score: '+str(score))
    plt.plot(x,y_pred,color='orange')
    plt.show()
```
